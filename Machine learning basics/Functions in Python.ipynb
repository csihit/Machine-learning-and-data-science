**Working With Functions in Python**

```python
#Defining a Function
def greet():
    print("Hello, welcome to Python!")

# Calling the function
greet()
```

```python
#Function Parameters
def greet_name(name):
    print(f"Hello, {name}!")

# Calling the function with an argument
greet_name("CSI")
```

```python
#Default Parameters
def greet_with_default(name="Guest"):
    print(f"Hello, {name}!")

# Calling the function without an argument
greet_with_default()
```

```python
#Returning Values
def square(x):
    return x ** 2

# Calling the function and storing the return value
result = square(5)
print("Square:", result)
```

```python
#Multiple Return Values
def divide(a, b):
    quotient = a // b
    remainder = a % b
    return quotient, remainder

# Calling the function and unpacking the return values
q, r = divide(10, 3)
print("Quotient:", q)
print("Remainder:", r)
```

```python
#Keyword Arguments
def person_info(name, age):
    print(f"Name: {name}, Age: {age}")

# Calling the function with keyword arguments
person_info(age=30, name="Alice")
```

```python
#Arbitrary Number of Arguments
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result

# Calling the function with multiple arguments
product = multiply(2, 3, 4)
print("Product:", product)
```

```python
#Lambda Functions (Anonymous Functions)
square_lambda = lambda x: x ** 2

# Calling the lambda function
result_lambda = square_lambda(3)
print("Square (Lambda):", result_lambda)
```

```python
#Recursive Functions
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Calling the recursive function
factorial_result = factorial(5)
print("Factorial:", factorial_result)
```

**Types of Functions in Python**

```python
#Built-in Functions
print(len([1, 2, 3]))  # Output: 3
print(max(5, 10, 2))   # Output: 10
```

```python
#User-Defined Functions
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")  # Output: Hello, Alice!
```

```python
#Anonymous Functions (Lambda Functions)
square = lambda x: x ** 2
print(square(5))  # Output: 25
```

```python
#Recursive Functions
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))  # Output: 120
```

```python
#Higher-Order Functions
def apply_operation(operation, x, y):
    return operation(x, y)

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

result_add = apply_operation(add, 5, 3)  # Output: 8
result_subtract = apply_operation(subtract, 10, 4)  # Output: 6
```

```python
#Generator Functions
def fibonacci(limit):
    a, b = 0, 1
    while a < limit:
        yield a
        a, b = b, a + b

fib_sequence = fibonacci(10)
for num in fib_sequence:
    print(num)  # Outputs Fibonacci sequence up to 10
```
