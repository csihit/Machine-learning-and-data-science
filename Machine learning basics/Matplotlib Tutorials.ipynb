**Matplotlib :**
*   **Useful for Making Plots**


```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# importing numpy
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
# plotting graphs
plt.plot(x, y)  # sine wave

# adding titles of x-axis and y-axis labels
plt.title('sine wave')
plt.xlabel('angle')
plt.ylabel('sine value')

plt.show()
```

```python
# plotting graphs
plt.plot(x, z)  # cosine wave

# adding titles of x-axis and y-axis labels
plt.title('cosine wave')
plt.xlabel('angle')
plt.ylabel('cosine value')

plt.show()
```

```python
# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [25, 30, 35, 40]
plt.figure(figsize=(6, 4))
plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(6, 4))
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Pie chart
sizes = [30, 20, 10, 40]
labels = ['A', 'B', 'C', 'D']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

**Generate 3D plots using Matplotlib**

```python
# Generate random 3D data points
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Create a figure and a set of subplots
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(x, y, z, c='b', marker='o')
# c='b': This specifies the color of the markers in the scatter plot. Here, 'b' stands for blue. You can use different color representations such as 'r' for red, 'g' for green, etc.
# marker='o': This specifies the shape of the markers used in the scatter plot. In this case, 'o' represents circular markers. Matplotlib supports various marker styles such as squares ('s'), triangles ('^'), crosses ('+'), etc.

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot')

plt.show()
```

**Matplotlib using diabetes.csv file**

```python
# importing pandas library
import pandas as pd
```

```python
# Load the diabetes dataset (assuming it's in CSV format and named 'diabetes.csv')
diabetes_df = pd.read_csv('/content/diabetes.csv')
```

```python
# Display the first few rows of the dataset to understand its structure
diabetes_df.head()
```

**Plotting examples**

```python
# Histogram of Age
plt.figure(figsize=(8, 6))
plt.hist(diabetes_df['Age'], bins=20, edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of Age vs. Blood Pressure
plt.figure(figsize=(8, 6))
plt.scatter(diabetes_df['Age'], diabetes_df['BloodPressure'])
plt.title('Scatter Plot: Age vs. Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.show()
```

```python
# 3. Box plot of BMI
plt.figure(figsize=(8, 6))
plt.boxplot(diabetes_df['BMI'])
plt.title('Box Plot: Body Mass Index (BMI)')
plt.ylabel('BMI')
plt.show()
```

```python
# Line plot of Diabetes Pedigree Function
plt.figure(figsize=(8, 6))
plt.plot(diabetes_df['DiabetesPedigreeFunction'])
plt.title('Line Plot: Diabetes Pedigree Function')
plt.xlabel('Patient Index')
plt.ylabel('Diabetes Pedigree Function')
plt.show()
```

```python
# Pie chart of Outcome
outcome_counts = diabetes_df['Outcome'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(outcome_counts, labels=['Non-Diabetic', 'Diabetic'], autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart: Outcome')
plt.show()
```
