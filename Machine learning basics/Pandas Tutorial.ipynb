**Data Manipulation**

```python
# Filtering rows based on conditions
filtered_df = diabetes_df[diabetes_df['BloodPressure'] > 70]
print(filtered_df)
```

```python
# Adding a new column
diabetes_df['BMI_category'] = pd.cut(diabetes_df['BMI'], bins=[0, 18.5, 24.9, 29.9, 100], labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
print(diabetes_df)
```

```python
diabetes_df.head()
```

```python
# Grouping and aggregating data
grouped_df = diabetes_df.groupby('BMI_category').mean()
print(grouped_df)
```

**Handling Missing Data**

```python
# Checking for missing values
print(diabetes_df.isnull().sum())
```

```python
# Dropping rows with missing values
diabetes_df.dropna(inplace=True)
```

**Correlation Of DataFrame**

```python
# Calculate the correlation matrix
correlation_matrix = housing.corr()
```

```python
housing.corr()
```

```python
# Filter positive correlations
correlation_matrix = housing.corr()
positive_correlations = correlation_matrix[correlation_matrix > 0]
print("Positive Correlations:")
print(positive_correlations)
```

```python
# Filter negative correlations
negative_correlations = correlation_matrix[correlation_matrix < 0]
print("\nNegative Correlations:")
print(negative_correlations)
```
