**Seaborn**

*   **Data Visualization Library**

```python
# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Set the style of the plots
sns.set_style("whitegrid")
```

```python
# 1. Distribution Plot (Histogram and KDE) for Total Bill
plt.figure(figsize=(8, 6))
sns.histplot(tips['total_bill'], kde=True, bins=20, color='blue')
plt.title("Distribution of Total Bill Amount")
plt.xlabel("Total Bill Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Scatter Plot of Total Bill vs. Tip
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="time")
plt.title("Scatter Plot: Total Bill vs. Tip")
plt.xlabel("Total Bill Amount")
plt.ylabel("Tip Amount")
plt.legend(title="Sex")
plt.show()
```

```python
# 3. Box Plot of Total Bill by Day and Time
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", hue="time", data=tips)
plt.title("Box Plot: Total Bill by Day and Time")
plt.xlabel("Day of Week")
plt.ylabel("Total Bill Amount")
plt.legend(title="Time")
plt.show()
```

```python
# 4. Violin Plot of Tip by Day and Sex
plt.figure(figsize=(8, 6))
sns.violinplot(x="day", y="tip", hue="sex", data=tips, split=True)
plt.title("Violin Plot: Tip by Day and Sex")
plt.xlabel("Day of Week")
plt.ylabel("Tip Amount")
plt.legend(title="Sex")
plt.show()
```

```python
# 5. Pair Plot (for numerical variables)
sns.pairplot(tips, hue="sex", markers=["o", "s"], palette="husl")
plt.suptitle("Pair Plot of Numerical Variables by Sex")
plt.show()
```

```python
# 6. Heatmap (correlation matrix)

# Filter out non-numeric columns (e.g., categorical columns) from the DataFrame
numerical_cols = tips.select_dtypes(include=[np.number])

# Calculate the correlation matrix for numerical columns only
correlation_matrix = numerical_cols.corr()

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

**Seaborn Library Funtions Using diabetes.csv Dataset**

```python
# Load the diabetes dataset (assuming it's in CSV format and named 'diabetes.csv')
diabetes_df = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_df.head()
```

```python
# Set the style of the plots
sns.set_style("whitegrid")
```

```python
# 1. Distribution Plot (Histogram and KDE) for Glucose
plt.figure(figsize=(8, 6))
sns.histplot(diabetes_df['Glucose'], kde=True, bins=20, color='blue')
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Scatter Plot of Glucose vs. Blood Pressure
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Glucose", y="BloodPressure", data=diabetes_df, hue="Outcome", style="Outcome")
plt.title("Scatter Plot: Glucose vs. Blood Pressure")
plt.xlabel("Glucose Level")
plt.ylabel("Blood Pressure")
plt.legend(title="Outcome")
plt.show()
```

```python
# 3. Box Plot of BMI by Outcome
plt.figure(figsize=(8, 6))
sns.boxplot(x="Outcome", y="BMI", data=diabetes_df)
plt.title("Box Plot: BMI by Outcome")
plt.xlabel("Outcome")
plt.ylabel("BMI")
plt.show()
```

```python
# 4. Violin Plot of BMI by Outcome and Age
plt.figure(figsize=(10, 6))
sns.violinplot(x='Outcome', y='BMI', hue='Age', data=diabetes_df, split=True)
plt.title("Violin Plot: BMI by Outcome and Age")
plt.xlabel("Outcome")
plt.ylabel("BMI")
plt.legend(title="Age")
plt.show()
```

```python
# 5. Count Plot of Outcome
plt.figure(figsize=(8, 6))
sns.countplot(x='Outcome', data=diabetes_df, palette='pastel')
plt.title("Count Plot of Outcome")
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.show()
```

```python
# 6. Bar Plot of Average Glucose by Outcome
average_glucose = diabetes_df.groupby('Outcome')['Glucose'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='Outcome', y='Glucose', data=average_glucose, palette='Set2')
plt.title("Bar Plot: Average Glucose by Outcome")
plt.xlabel("Outcome")
plt.ylabel("Average Glucose")
plt.show()
```

```python
# 8. Pair Plot with Hue by Outcome
sns.pairplot(diabetes_df, hue='Outcome', markers=['o', 's'])
plt.suptitle("Pair Plot by Outcome")
plt.show()
```

```python
# 9. Correlation Heatmap
correlation_matrix = diabetes_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()
```
