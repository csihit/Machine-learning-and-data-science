**Imbalanced Dataset:**

**A dataset with an unequal class distribution.**

```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
# loading the dataset into pandas DataFrame
credit_card_data = pd.read_csv("/content/credit_data.csv")
```

```python
credit_card_data.head()
```

```python
credit_card_data.shape
```

```python
# distribution of the 2 classes
credit_card_data['Class'].value_counts()
```

**0 --> Legit Transactions**

**1 --> fraudulent Transactions**

```python
# separating the legit and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

**Under-sampling**

**Build a sample dataset containing similar distribution of Legit & Fraudulent Transactions**

**Number of Fraudulent Transactions --> 492**

```python
legit_sample = legit.sample(n = 492)
```

```python
legit_sample.shape
```

**Concatenating the 2 DataFrames**

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```
