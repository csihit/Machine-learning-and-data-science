```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate synthetic data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # Independent variable (1 feature)
y = 4 + 3 * X + np.random.randn(100, 1)  # Dependent variable (target)
```

```python
X
```

```python
y
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(X, y)
```

```python
# Make predictions
X_new = np.array([[0], [2]])  # New data points for prediction
y_pred = model.predict(X_new)
```

```python
# Plot the data and the linear regression line
plt.scatter(X, y, label='Data Points')
plt.plot(X_new, y_pred, color='red', label='Linear Regression Line')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```

```python
# Print the coefficients (slope and intercept)
print(f"Slope (Coefficient): {model.coef_[0][0]:.2f}")
print(f"Intercept: {model.intercept_[0]:.2f}")
```
