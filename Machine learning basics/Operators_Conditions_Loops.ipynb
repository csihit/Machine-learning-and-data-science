**Operators in Python**

```python
# Arithmetic operators
a = 10
b = 5
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Exponentiation:", a ** b)
print("Modulus:", a % b)
```

```python
# Comparison operators
x = 10
y = 20
print("Equal to:", x == y)
print("Not equal to:", x != y)
print("Greater than:", x > y)
print("Less than:", x < y)
print("Greater than or equal to:", x >= y)
print("Less than or equal to:", x <= y)
```

```python
# Logical operators
p = True
q = False
print("Logical AND:", p and q)
print("Logical OR:", p or q)
print("Logical NOT:", not p)
```

```python
# Assignment operators
total = 100
total += 50  # Equivalent to total = total + 50
print("Total after addition:", total)
total -= 30  # Equivalent to total = total - 30
print("Total after subtraction:", total)
total *= 2  # Equivalent to total = total * 2
print("Total after multiplication:", total)
total /= 3  # Equivalent to total = total / 3
print("Total after division:", total)
total %= 4  # Equivalent to total = total % 4
print("Total after modulus:", total)
```

```python
# Identity operators
list1 = [1, 2, 3]
list2 = [1, 2, 3]
print("Is list1 same as list2 (using 'is'):", list1 is list2)
print("Is list1 same as list2 (using '=='):", list1 == list2)
```

```python
# Membership operators
fruits = ['apple', 'banana', 'cherry']
print("Is 'banana' in fruits:", 'banana' in fruits)
print("Is 'grape' not in fruits:", 'grape' not in fruits)
```

```python
# Bitwise operators
num1 = 10  # 1010 in binary
num2 = 6   # 0110 in binary
print("Bitwise AND:", num1 & num2)  # 0010 (2 in decimal)
print("Bitwise OR:", num1 | num2)   # 1110 (14 in decimal)
print("Bitwise XOR:", num1 ^ num2)  # 1100 (12 in decimal)
print("Bitwise NOT:", ~num1)        # -11 (complement of 10 in binary)
```

**Conditional Statements in Python**

```python
# If statement
x = 10
if x > 5:
    print("x is greater than 5")
```

```python
# If-else statement
y = 3
if y % 2 == 0:
    print("y is even")
else:
    print("y is odd")
```

```python
# If-elif-else statement
z = -5
if z > 0:
    print("z is positive")
elif z == 0:
    print("z is zero")
else:
    print("z is negative")
```

```python
# Nested if statements
a = 15
b = 20
if a > 10:
    if b > 15:
        print("Both a and b are greater than their respective thresholds")
```

```python
# Ternary conditional operator (Conditional Expression)
result = "Even" if a % 2 == 0 else "Odd"
print("Result:", result)
```

```python
# Short-circuit evaluation
value = None
if value is not None and len(value) > 0:
    print("Value is not empty")
else:
    print("Value is None or empty")
```

**Loops In Python**

```python
# Iterating over a list
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)

# Using range() function
for i in range(5):
    print(i)
```

```python
# Using a while loop to print numbers from 1 to 5
num = 1
while num <= 5:
    print(num)
    num += 1
```

```python
# Using break to exit the loop early
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# Using continue to skip the current iteration
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# Using else with a for loop
for i in range(3):
    print(i)
else:
    print("Loop completed without a break")

# Using else with a while loop
num = 1
while num <= 5:
    print(num)
    num += 1
else:
    print("While loop completed")
```

```python
# Nested for loops
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
# Nested while loops
outer = 1
while outer <= 2:
    inner = 1
    while inner <= 3:
        print(outer, inner)
        inner += 1
    outer += 1
```

```python
# Using pass as a placeholder
for i in range(5):
    if i == 2:
        pass
    else:
        print(i)
```
